```python
# Analiza widma sygnału sinusoidalnego
#
import numpy as np
import matplotlib.pyplot as plt

# Parametry sygnału
Fs = 1000  # Częstotliwość próbkowania
T = 1.0 / Fs  # Okres próbkowania
t = np.arange(0, 1, T)  # Wektor czasu
f = 5  # Częstotliwość sygnału

# Generowanie sygnału sinusoidalnego
signal = np.sin(2 * np.pi * f * t)

# Obliczanie transformaty Fouriera
N = len(signal)
fft_signal = np.fft.fft(signal)
freqs = np.fft.fftfreq(N, T)

# Wizualizacja widma sygnału
plt.figure(figsize=(10, 6))
plt.plot(freqs[:N//2], np.abs(fft_signal[:N//2]))
plt.title('Widmo sygnału sinusoidalnego')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()
plt.show()
```

```python
# Filtracja sygnału za pomocą filtru dolnoprzepustowego
#
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Parametry sygnału
Fs = 1000  # Częstotliwość próbkowania
T = 1.0 / Fs  # Okres próbkowania
t = np.arange(0, 1, T)  # Wektor czasu
f = 5  # Częstotliwość sygnału

# Generowanie sygnału zaszumionego
signal = np.sin(2 * np.pi * f * t)
noise = np.random.normal(0, 0.5, signal.shape)
noisy_signal = signal + noise

# Projektowanie filtra dolnoprzepustowego
cutoff = 10  # Częstotliwość graniczna
b, a = butter(4, cutoff / (0.5 * Fs), btype='low')

# Filtracja sygnału
filtered_signal = filtfilt(b, a, noisy_signal)

# Wizualizacja sygnału przed i po filtracji
plt.figure(figsize=(10, 6))
plt.plot(t, noisy_signal, label='Zaszumiony sygnał')
plt.plot(t, filtered_signal, label='Sygnał po filtracji', linewidth=2)
plt.title('Filtracja sygnału za pomocą filtra dolnoprzepustowego')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.legend()
plt.grid()
plt.show()
```

```python
# Analiza widma sygnału z modulowaną amplitudą
#
import numpy as np
import matplotlib.pyplot as plt

# Parametry sygnału
Fs = 1000  # Częstotliwość próbkowania
T = 1.0 / Fs  # Okres próbkowania
t = np.arange(0, 1, T)  # Wektor czasu
fc = 50  # Częstotliwość nośna
fm = 5  # Częstotliwość modulująca

# Generowanie sygnału modulowanego AM
message_signal = np.sin(2 * np.pi * fm * t)
carrier_signal = np.sin(2 * np.pi * fc * t)
am_signal = (1 + message_signal) * carrier_signal

# Obliczanie transformaty Fouriera
N = len(am_signal)
fft_signal = np.fft.fft(am_signal)
freqs = np.fft.fftfreq(N, T)

# Wizualizacja widma sygnału
plt.figure(figsize=(10, 6))
plt.plot(freqs[:N//2], np.abs(fft_signal[:N//2]))
plt.title('Widmo sygnału modulowanego AM')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()
plt.show()
```

```python
# Analiza widma sygnału kwadratowego
#
import numpy as np
import matplotlib.pyplot as plt

# Parametry sygnału
Fs = 1000  # Częstotliwość próbkowania
T = 1.0 / Fs  # Okres próbkowania
t = np.arange(0, 1, T)  # Wektor czasu
f = 5  # Częstotliwość sygnału

# Generowanie sygnału kwadratowego
square_signal = np.sign(np.sin(2 * np.pi * f * t))

# Obliczanie transformaty Fouriera
N = len(square_signal)
fft_signal = np.fft.fft(square_signal)
freqs = np.fft.fftfreq(N, T)

# Wizualizacja widma sygnału
plt.figure(figsize=(10, 6))
plt.plot(freqs[:N//2], np.abs(fft_signal[:N//2]))
plt.title('Widmo sygnału kwadratowego')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()
plt.show()
```

```python
# Okienkowanie sygnału
#
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import windows

# Parametry sygnału
Fs = 1000  # Częstotliwość próbkowania
T = 1.0 / Fs  # Okres próbkowania
t = np.arange(0, 1, T)  # Wektor czasu
f = 5  # Częstotliwość sygnału

# Generowanie sygnału
signal = np.sin(2 * np.pi * f * t)

# Zastosowanie okien
hann_window = windows.hann(len(signal))
hamming_window = windows.hamming(len(signal))

# Okienkowanie sygnału
hann_signal = signal * hann_window
hamming_signal = signal * hamming_window

# Obliczanie transformaty Fouriera
N = len(signal)
fft_signal = np.fft.fft(signal)
fft_hann_signal = np.fft.fft(hann_signal)
fft_hamming_signal = np.fft.fft(hamming_signal)
freqs = np.fft.fftfreq(N, T)

# Wizualizacja widma sygnału
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(freqs[:N//2], np.abs(fft_signal[:N//2]))
plt.title('Widmo sygnału przed okienkowaniem')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(freqs[:N//2], np.abs(fft_hann_signal[:N//2]))
plt.title('Widmo sygnału po okienkowaniu oknem Hanninga')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(freqs[:N//2], np.abs(fft_hamming_signal[:N//2]))
plt.title('Widmo sygnału po okienkowaniu oknem Hamminga')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Ilustracja twierdzenia Nyquista
#
import numpy as np
import matplotlib.pyplot as plt

# Parametry sygnału
Fs = 1000  # Częstotliwość próbkowania
T = 1.0 / Fs  # Okres próbkowania
t = np.arange(0, 1, T)  # Wektor czasu
f = 50  # Częstotliwość sygnału

# Generowanie sygnału
signal = np.sin(2 * np.pi * f * t)

# Próbkowanie sygnału z różnymi częstotliwościami próbkowania
Fs_low = 80  # Częstotliwość próbkowania niższa niż 2 * f
Fs_high = 200  # Częstotliwość próbkowania wyższa niż 2 * f

t_low = np.arange(0, 1, 1.0 / Fs_low)
t_high = np.arange(0, 1, 1.0 / Fs_high)

signal_low = np.sin(2 * np.pi * f * t_low)
signal_high = np.sin(2 * np.pi * f * t_high)

# Obliczanie transformaty Fouriera
N_low = len(signal_low)
N_high = len(signal_high)
fft_signal_low = np.fft.fft(signal_low)
fft_signal_high = np.fft.fft(signal_high)
freqs_low = np.fft.fftfreq(N_low, 1.0 / Fs_low)
freqs_high = np.fft.fftfreq(N_high, 1.0 / Fs_high)

# Wizualizacja sygnału przed i po próbkowaniu
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal, label='Oryginalny sygnał')
plt.plot(t_low, signal_low, 'o', label='Próbkowany sygnał (Fs = 80 Hz)')
plt.title('Sygnał przed i po próbkowaniu')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(t, signal, label='Oryginalny sygnał')
plt.plot(t_high, signal_high, 'o', label='Próbkowany sygnał (Fs = 200 Hz)')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# Wizualizacja widma sygnału przed i po próbkowaniu
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(freqs_low[:N_low//2], np.abs(fft_signal_low[:N_low//2]))
plt.title('Widmo sygnału po próbkowaniu (Fs = 80 Hz)')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(freqs_high[:N_high//2], np.abs(fft_signal_high[:N_high//2]))
plt.title('Widmo sygnału po próbkowaniu (Fs = 200 Hz)')
plt.xlabel('Częstotliwość [Hz]')
plt.ylabel('Amplituda')
plt.grid()

plt.tight_layout()
plt.show()
```
